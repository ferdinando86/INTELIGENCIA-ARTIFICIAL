```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split 
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor 
from sklearn import metrics
from IPython.display import Image  
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes(return_X_y=False)
diabetes.data
```

```python
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.3, random_state=1)
pd.DataFrame(X_train).head()
```

### Definir y entrenar regresor

```python
mlp = MLPRegressor(max_iter=5000,
                   hidden_layer_sizes=(55),
                   activation='logistic',
                   learning_rate_init=0.3,
                   verbose = True
                   )
mlp.fit(X_train, y_train)
```

### Evaluar error obtenido

```python
y_pred = mlp.predict(X_test)
error = metrics.mean_squared_error(y_test,y_pred)
print(error)
```

En la mayoria de los casos podemos observar que el valor de la prediccin varia mucho respecto al valor real de la progresin de la enfermedad.
Esto significa para el paciente que las predicciones no son certeras y el sistema no es confiable para utilizarse en nuevos pacientes.


## Entrenar modelo

```python
reg = DecisionTreeRegressor(max_depth=2)
reg = reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

## Calcular el error

```python
error = metrics.mean_squared_error(y_test, y_pred)
print(error)
```

## Predicciones con ambos mtodos

```python
y_predmlp = []
y_predreg = []
y_sample = []
for i in range(0,10):
    aux = random.randrange(0, len(X_test)-1)
    X_sample = X_test.values[aux]
    y_sample.append(y_test.values[aux])
    y_predmlp.append(mlp.predict(X_sample.reshape(1, -1)))
    y_predreg.append(reg.predict(X_sample.reshape(1, -1)))
    print("La progresin de la enfermedad segn red neuronal en un ao es de "+str(y_predmlp[i]))
    print("La progresin de la enfermedad segn arbol de decisin en un ao es de "+str(y_predreg[i]))
    print("La progresin real de la enfermedad es "+str(y_sample[i])+"\n")
```

```python
plt.scatter(y_predmlp, y_sample, color = 'blue')
plt.scatter(y_predreg, y_sample, color = 'red')
plt.xlabel('Prediccin')
plt.ylabel('Valor real')
plt.plot([0,400],[0,400], 'black', linewidth=1)
plt.show()
```

Segn el grfico de predicciones y valores reales, las predicciones del rbol de decisin son mejores que las de la red neuronal.
